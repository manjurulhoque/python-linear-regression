```python
df = pd.read_csv("Housing.csv")
```

```python
Y = df['price']
X = df['lotsize']
 
X=X.values.reshape(len(X),1)
Y=Y.values.reshape(len(Y),1)
```

```python
X_train = X[:-250]
X_test = X[-250:]
```

```python
Y_train = Y[:-250]
Y_test = Y[-250:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(X_test, Y_test,  color='black')
plt.title('Test Data')
plt.xlabel('Size')
plt.ylabel('Price')
plt.xticks(())
plt.yticks(())
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(X_train, Y_train)
```

```python
plt.scatter(X_test, Y_test,  color='black')
plt.title('Linear Regression')
plt.xlabel('Size')
plt.ylabel('Price')
plt.xticks(())
plt.yticks(())
plt.plot(X_test, regr.predict(X_test), color='yellow', linewidth=3)# best fit
```
